```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv("heart.csv")
```

```python
ohe = OneHotEncoder()
```

```python
print(ohe)
```

```python
ohe.fit_transform(df[["Gender", "ChestPainType", "RestingECG"]]).toarray()
```

```python
#saved array result
feature_array = ohe.fit_transform(df[["Gender", "ChestPainType", "RestingECG"]]).toarray()
```

```python
print(feature_array)
```

```python
ohe.categories_
```

```python
#saved categories as feature_labels
feature_labels = ohe.categories_
```

```python
#convert arrays in one array
combined_array = np.concatenate((feature_labels))
```

```python
print(combined_array)
```

```python
pd.DataFrame(feature_array, columns = combined_array)
```

```python
#saved results in features
features = pd.DataFrame(feature_array, columns = combined_array)
```

```python
print(features)
```

```python
features.head()
```

```python
pd.concat([df, features], axis=1)
```

```python
#real data with one hot encoding data in df_new
df_new = pd.concat([df, features], axis=1)
```
